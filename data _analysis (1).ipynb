```python
crime = pd.read_csv('crime.2020.csv', engine='python', encoding='euc-kr')
```

```python
# 조건 설정
cond1 = (crime['범죄대분류'] == '강력범죄')
crime2 = crime.loc[cond1].drop(columns = '범죄대분류')
crime2
```

```python
# 범죄중분류별 합계 계산
crime2['합계'] = np.sum(crime2, axis=1)
crime2

# 합계를 기준으로 내림차순
crime3 = crime2[['범죄중분류','합계']].sort_values(by='합계', ascending = False)
crime3
```

```python
#강력범죄의 범죄중분류 건수
plt.figure(figsize=(10, 5))
plt.rc('font', family='Malgun Gothic') # 힌글 폰트 지정(맑은 고딕)
plt.title('강력범죄의 범죄중분류 건수', fontsize = 20)

sns.barplot(data=crime3, y='범죄중분류', x='합계', palette='Set1')  # 팔레트 색상 지정


plt.xlabel('빈도수', fontsize=15)# x축 제목 서식 설정
plt.ylabel('범죄 종류', fontsize=15)# y축 제목 서식 설정
plt.xticks(fontsize=14)# x축 레이블 서식 설정
plt.yticks(fontsize=14, rotation=10)# y축 레이블 서식 설정

# 그래프 위에 값 표시
for i, v in enumerate(crime3['합계']):
    plt.text(v + 3, i, str(v), color='black', ha='left', va='center', fontsize=10)

plt.tight_layout()  # 그래프 요소 간격 조정

plt.show()
## 해당 그래프는 x, y가 반대이므로 x, y값 서식 설정을 반대로 생각해서 지정해야 합니다.
```

```python
crime_violence = crime[crime['범죄대분류'] == '폭력범죄']
```

```python
crime_violence
```

```python
# 조건 설정
cond2 = (crime_violence['범죄대분류'] == '폭력범죄')

# 폭력범죄의 종류와 합계 출력
crime3 = crime_violence[cond1].drop(columns='범죄대분류')

print(crime3)
```

```python
# 범죄중분류별 합계 계산
crime3['합계'] = np.sum(crime2, axis=1)
crime3
# 합계를 기준으로 내림차순
crime4 = crime3[['범죄중분류','합계']].sort_values(by='합계', ascending = False)
crime4
```

```python
plt.figure(figsize=(10, 5))
plt.rc('font', family='Malgun Gothic')
plt.title('폭력범죄의 범죄중분류 건수', fontsize=20)

# Stripplot으로 데이터 시각화
sns.stripplot(data=crime4, y='범죄중분류', x='합계', palette='Set1', size=8)

plt.xlabel('빈도수', fontsize=15)
plt.ylabel('범죄 종류', fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 그래프 위에 값 표시
for i, v in enumerate(crime4['합계']):
    plt.text(v + 3, i, str(v), color='black', ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()

```

```python
crime_intelligence = crime[crime['범죄대분류'] == '지능범죄']
```

```python
crime_intelligence
```

```python
# 조건 설정
cond3= (crime_intelligence['범죄대분류'] == '지능범죄')

# 지능범죄의 종류와 합계 출력
crime4 = crime_intelligence[cond3].drop(columns='범죄대분류')

print(crime4)
```

```python
# 범죄중분류별 합계 계산
crime4['합계'] = np.sum(crime4, axis=1)
crime4
# 합계를 기준으로 내림차순
crime5 = crime4[['범죄중분류','합계']].sort_values(by='합계', ascending = False)
crime5
```

```python
# 숫자 크기를 천 단위로 표기하는 함수 정의
def format_thousands(x, pos):
    return f'{x/1000:.0f}천'

#지능범죄의 범죄중분류 건수
# 그래프 설정
plt.figure(figsize=(10, 5))
plt.title('지능범죄의 범죄중분류 건수', fontsize=20)

# 로그 스케일 적용하여 그래프 그리기
sns.barplot(data=crime5, y='범죄중분류', x='합계', palette='Set1', log=True)

plt.xlabel('빈도수', fontsize=15)
plt.ylabel('범죄 종류', fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# x축 숫자 크기를 천 단위로 표기
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_thousands))


plt.tight_layout()
plt.show()
```

```python
# 혼인 합계, 미혼 합계 컬럼 생성
crime['혼인_합계'] = crime['혼인관계(유배우자)'] + crime['혼인관계(동거)'] + crime['혼인관계(이혼)'] + crime['혼인관계(사별)']

crime['미혼_합계'] = crime['미혼자부모관계(실(양)부모)'] + crime['미혼자부모관계(계부모)'] + crime['미혼자부모관계(실부계모)'] + crime['미혼자부모관계(실부무모)'] + crime['미혼자부모관계(실모계부)'] + crime['미혼자부모관계(실모무부)'] + crime['미혼자부모관계(계부무모)'] + crime['미혼자부모관계(계모무부)'] + crime['미혼자부모관계(무부모)']

crime.head()
```

```python
crime6 = crime[['범죄중분류', '혼인_합계', '미혼_합계']].set_index('범죄중분류')
crime6.sort_values(by='혼인_합계', ascending = False, inplace = True)
crime6
```

```python

#  새로운 색상 팔레트 정의
colors = ['#FF6F61', '#6B5B95']

# 막대 그래프 시각화
ax = crime6.plot(kind='bar', title='범죄분석', figsize=(16, 8), legend=True, fontsize=12, color=colors, edgecolor='black')

# 그림자 추가
ax.set_box_aspect(0.5)
ax.set_facecolor('white')
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge')

# x축, y축 정보 표시
ax.set_xlabel('범죄중분류', fontsize=12)
ax.set_ylabel('범죄건수', fontsize=12)

# 범례 위치 조정
ax.legend(['혼인_합계', '미혼_합계'], fontsize=12, loc='upper right')

# 그리드 추가
ax.grid(axis='y', linestyle='--')

# 막대 그래프 위에 값 표시
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', color='black', fontsize=10)


# 눈금선 및 타이틀 제거
ax.tick_params(axis='x', bottom=False, labelbottom=False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.show()
```
